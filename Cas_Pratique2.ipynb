```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
E=pd.ExcelFile("examen1.xlsx")
Ex=E.parse("Resultats")
Ex
```

```python
Phy=Ex.iloc[:,2:5]
moyPhy=((Phy.Physique1*2.5)+(Phy.Physique2*2.5)+Phy.TPPhysique)/6
StatPhy=[]
for i in range(44):
    if moyPhy[i]<10:
        StatPhy.append('NV')
    else:
        StatPhy.append('V')
StatPhy    
```

```python
Math=Ex.iloc[:,5:8]
moyMat=((Math.Maths1*2.5)+(Math.Maths2*2.5))/5
StatMat=[]
for i in range(44):
    if moyMat[i]<10:
        StatMat.append('NV')
    else:
        StatMat.append('V')
StatMat    
```

```python
Chi=Ex.iloc[:,7:10]
moyChi=((Chi.Chimie1*2.5)+(Chi.Chimie2*2.5)+Chi.TPChimie)/6
StatChi=[]
for i in range(44):
    if moyChi[i]<10:
        StatChi.append('NV')
    else:
        StatChi.append('V')
StatChi   
```

```python
Inf=Ex.iloc[:,10:12]
moyInf=((Inf.Info1*2.5)+(Inf.Info2*2.5))/5
StatInf=[]
for i in range(44):
    if moyInf[i]<10:
        StatInf.append('NV')
    else:
        StatInf.append('V')
StatInf   
```

```python
Base=Ex.iloc[:,12:16]
moyBase=((Base.Base1*1.5)+(Base.Base2*1.5)+Base.Base3)/4
StatBase=[]
for i in range(44):
    if moyBase[i]<10:
        StatBase.append('NV')
    else:
        StatBase.append('V')
StatBase   
```

```python
Graph=Ex.iloc[:,15:18]
moyGraph=((Graph.Graphique1*2)+(Graph.Graphique2*2))/4
StatGraph=[]
for i in range(44):
    if moyGraph[i]<10:
        StatGraph.append('NV')
    else:
        StatGraph.append('V')
StatGraph   
```

```python
tri_Ex=Ex.sort_values(by = ["Nom","Prénom"],ascending = ([True,True]))
tri_Ex
```

```python
tri_Ex.insert(5,"Statut_Physique",StatPhy,allow_duplicates=False)
tri_Ex.insert(8,"Statut_Maths",StatMat,allow_duplicates=False)
tri_Ex.insert(12,"Statut_Chimie",StatChi,allow_duplicates=False)
tri_Ex.insert(15,"Statut_Informatique",StatInf,allow_duplicates=False)
tri_Ex.insert(19,"Statut_Base",StatBase,allow_duplicates=False)
tri_Ex.insert(22,"Statut_Graphique",StatGraph,allow_duplicates=False)
```

```python
#tri_Ex
```

```python
Resultats=pd.ExcelWriter("Resultats.xlsx",engine="xlsxwriter")
tri_Ex.to_excel(Resultats,"Deliberation",index=False)#Creation d'une feuille dans le fichier
Resultats.save()
```

```python
moy=(6*moyPhy+5*moyMat+5*moyInf+6*moyChi+4*moyBase+4*moyGraph)/30
j=np.where(moy==moy.max())
print("Le major de la classe est", list(Ex.Prénom[j[0]])[0], list(Ex.Nom[j[0]])[0])
```

```python
Val_Phy=tri_Ex.Statut_Physique.value_counts()[0]
Val_Mat=tri_Ex.Statut_Maths.value_counts()[0]
Val_Chi=tri_Ex.Statut_Chimie.value_counts()[0]
Val_Inf=tri_Ex.Statut_Informatique.value_counts()[0]
Val_Base=tri_Ex.Statut_Base.value_counts()[0]
Val_Graph=tri_Ex.Statut_Graphique.value_counts()[0]
Val=[Val_Phy,Val_Mat,Val_Chi,Val_Inf,Val_Base,Val_Graph]
UE=['Physique','Maths','Chimie','Informatique','Base','Graphique']
```

```python
ax=plt.plot(UE,Val)
plt.grid()
plt.show()
```
